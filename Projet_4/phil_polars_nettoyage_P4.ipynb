```python
i = 1
for column in data_2016.select(pl.col(pl.NUMERIC_DTYPES)).get_columns():
    distribution = column.value_counts().sort(column.name)
    plt.figure(1, figsize=(16, 40), dpi=72)
    plt.subplot(9, 3, i)
    plt.title(column.name)
    plt.autoscale()
    plt.bar(x=distribution[column.name], height=distribution["counts"])
    i += 1
```

```python
data_2016_cleaned = data_2016 \
    .with_columns(
        pl.when((pl.col("SecondLargestPropertyUseType").is_null()) & (pl.col("SecondLargestPropertyUseTypeGFA").is_null()))
            .then(pl.lit(0))
            .otherwise(pl.col("SecondLargestPropertyUseTypeGFA"))
            .alias("SecondLargestPropertyUseTypeGFA")
    ) \
    .with_columns(
        pl.when((pl.col("ThirdLargestPropertyUseType").is_null()) & (pl.col("ThirdLargestPropertyUseTypeGFA").is_null()))
            .then(pl.lit(0))
            .otherwise(pl.col("ThirdLargestPropertyUseTypeGFA"))
            .alias("ThirdLargestPropertyUseTypeGFA")
    ) \
    .with_columns(pl.col("SecondLargestPropertyUseType").fill_null(pl.lit(""))) \
    .with_columns(pl.col("ThirdLargestPropertyUseType").fill_null(pl.lit("")))
```

```python
data_2016_cleaned \
    .filter((pl.col("ENERGYSTARScore").is_not_null())) \
    .null_count() \
    .sum() \
    .transpose() \
    .with_columns(pl.Series(name="name", values=data_2016.columns)) \
    .filter(pl.col("column_0") > 0) \
    .select(["name", "column_0"])
```

```python

data_2016_cleaned.filter((pl.col("NumberofBuildings").is_null()))
```

```python
column = "GFAbyBuilding"

distribution = data_2016_cleaned \
    .filter((pl.col("NumberofBuildings") > 1)) \
    .with_columns((pl.col("PropertyGFABuilding(s)") / pl.col("NumberofBuildings")).alias(column)) \
    .get_column(column)

sns.displot(distribution)


```

```python
column = "GFAbyBuilding"

distribution = data_2016_cleaned \
    .filter((pl.col("NumberofBuildings") > 1) & (pl.col("PropertyGFABuilding(s)") / pl.col("NumberofBuildings") > 200000))

distribution
```

```python
for property_type in data_2016_cleaned.select("PrimaryPropertyType").unique().sort("PrimaryPropertyType").get_column("PrimaryPropertyType"):
    distribution = data_2016_cleaned \
        .filter((pl.col("PrimaryPropertyType") == property_type) & (pl.col("ENERGYSTARScore").is_not_null())) \
        .with_columns((pl.col("PropertyGFABuilding(s)") / pl.col("NumberofBuildings")).alias("GFAbyBuilding")) \
        .get_column("GFAbyBuilding")
    sns.displot(distribution).set(title=property_type)
    
```
