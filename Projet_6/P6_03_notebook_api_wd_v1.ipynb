```python
import http.client
import json
import csv
import os

def connect_to_api():
    try:
        conn = http.client.HTTPSConnection("edamam-food-and-grocery-database.p.rapidapi.com")

        headers = {
            'X-RapidAPI-Key': "82e7f76641msh8efe42731f3e34bp1164ddjsn1d61cc9d0dc1",
            'X-RapidAPI-Host': "edamam-food-and-grocery-database.p.rapidapi.com"
        }

        # Vérification de la connexion à l'API
        conn.request("GET", "/api/food-database/v2/parser?nutrition-type=cooking&category%5B0%5D=generic-foods&health%5B0%5D=alcohol-free", headers=headers)
        
        res = conn.getresponse()
        if res.status != 200:
            print("Erreur de connexion à l'API.")
            return None, None

        print("Connexion à l'API réussie.")
        return conn, headers
    except Exception as e:
        print(f"Erreur lors de la connexion à l'API : {str(e)}")
        return None, None

def get_food_data(conn, headers, query):
    try:
        conn = http.client.HTTPSConnection("edamam-food-and-grocery-database.p.rapidapi.com")
        
        conn.request("GET", f"/api/food-database/v2/parser?{query}", headers=headers)
        
        res = conn.getresponse()
        if res.status != 200:
            print(f"Erreur lors de la récupération des données : {res.status} - {res.reason}")
            return None

        data = res.read().decode("utf-8")
        return json.loads(data)
    except Exception as e:
        print(f"Erreur lors de la récupération des données : {str(e)}")
        return None

def extract_food_items(data, num_items=100, filter_ingredient=""):
    try:
        items = data.get("hints", [])[:num_items]
        extracted_data = []

        for item in items:
            food_data = item.get("food", {})
            food_id = food_data.get("foodId")
            label = food_data.get("label")
            category = food_data.get("category")
            food_contents_label = food_data.get("foodContentsLabel")
            image = food_data.get("image")
            ingr = ", ".join(food_data.get("ingredients", []))

            # Filtrer par ingrédient
            if filter_ingredient is not None and filter_ingredient.lower() not in ingr.lower():
                continue

            extracted_data.append({
                "foodId": food_id,
                "label": label,
                "category": category,
                "foodContentsLabel": food_contents_label,
                "image": image,
                "ingr": ingr
            })

        return extracted_data
    except Exception as e:
        print(f"Erreur lors de l'extraction des données : {str(e)}")
        return []

def save_to_csv(data, filepath):
    try:
        with open(filepath, mode="w", newline="") as csv_file:
            fieldnames = ["foodId", "label", "category", "foodContentsLabel", "image", "ingr"]
            writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
            
            writer.writeheader()
            for item in data:
                writer.writerow(item)

        if os.path.isfile(filepath):
            print(f"Le fichier CSV a été enregistré avec succès sous le chemin : {filepath}")
        else:
            print("Une erreur s'est produite lors de l'enregistrement du fichier CSV.")
    except Exception as e:
        print(f"Erreur lors de l'enregistrement du fichier CSV : {str(e)}")

def main():
    conn, headers = connect_to_api()
    
    if conn is None or headers is None:
        return
    
    query = "nutrition-type=cooking&category%5B0%5D=generic-foods&health%5B0%5D=alcohol-free"
    data = get_food_data(conn, headers, query)
    
    if data is not None:
#         filter_ingredient = "champagne"  # Modifier ici l'ingrédient de filtrage
        filter_ingredient = ""
        extracted_data = extract_food_items(data, num_items=10, filter_ingredient=filter_ingredient)
        
        # Chemin d'enregistrement du fichier CSV (paramétrable)
        save_path = r'C:\Users\Zbook\OpenClassRoom\ProjetsGIT\Projet_6\data\food_data_v3.csv'
        
        save_to_csv(extracted_data, save_path)

if __name__ == "__main__":
    main()
```
