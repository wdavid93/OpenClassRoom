```python
def attaquer_village(village_depart, village_arrive, troupes_file):
    firefox.type_keys('^l')  # Presser la combinaison de touches Ctrl+L pour sélectionner la barre d'adresse
    url = f"https://fr81.guerretribale.fr/game.php?village={village_depart}&screen=place&target={village_arrive}"
    firefox.type_keys(url + '{ENTER}')  # Entrer l'URL et appuyer sur Entrée
    delay_with_random_seconds()
    
    # Charger les données des troupes à partir du fichier
    with open(troupes_file, 'r') as file:
        lines = file.readlines()
        headers = lines[0].strip().split(',')
        data = lines[1].strip().split(',')

    # Saisir les champs pour chaque type de troupe
    for header, value in zip(headers, data):
        zone = firefox.child_window(auto_id=header)
        zone.type_keys(str(value))
        delay_with_random_seconds()

    bouton = firefox.child_window(auto_id='target_attack')
    bouton.click()
    delay_with_random_seconds()

    bouton = firefox.child_window(auto_id='troop_confirm_submit')
    bouton.click()
    delay_with_random_seconds()
```
