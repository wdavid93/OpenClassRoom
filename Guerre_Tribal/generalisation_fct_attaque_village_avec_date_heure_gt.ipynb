```python
import csv
import datetime

def mettre_a_jour_heure_attaque():
    fichier_csv = "C:\\Users\\Zbook\\OpenClassRoom\\Projet\\Guerre_Tribal\\data\\villages.csv"

    # Vérifier si le fichier existe déjà
    fichier_existe = False
    try:
        with open(fichier_csv, 'r') as file:
            reader = csv.reader(file)
            headers = next(reader)
            if 'heure_attaque' in headers:
                fichier_existe = True
    except FileNotFoundError:
        pass

    # Si le fichier n'existe pas ou si la colonne 'heure_attaque' n'existe pas, créer la colonne
    if not fichier_existe:
        headers = []
        try:
            with open(fichier_csv, 'r') as file:
                reader = csv.reader(file)
                headers = next(reader)
        except FileNotFoundError:
            pass

        headers.append('heure_attaque')
        with open(fichier_csv, 'w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(headers)

    # Mettre à jour la colonne 'heure_attaque' avec la date et l'heure actuelles
    with open(fichier_csv, 'r') as file:
        reader = csv.DictReader(file)
        lignes = list(reader)
    print(lignes)
    for ligne in lignes:
        ligne['heure_attaque'] = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    with open(fichier_csv, 'w', newline='') as file:
        writer = csv.DictWriter(file, fieldnames=lignes[0].keys())
        writer.writeheader()
        writer.writerows(lignes)
```
