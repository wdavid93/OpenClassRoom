```python
from pyspark.sql import SparkSession
from pyspark.sql.types import StringType, BinaryType, StructType, StructField
from pyspark.sql import Row
import base64

spark = SparkSession.builder.getOrCreate()

spark.conf.set("spark.hadoop.fs.AbstractFileSystem.file.impl", "org.apache.hadoop.fs.LocalFileSystem")

# image_dir = "/chemin/vers/le/repertoire"  # Spécifiez le chemin vers votre répertoire d'images

# Lecture récursive de tous les fichiers binaires dans le répertoire
file_rdd = spark.sparkContext.binaryFiles(image_dir)

# Filtrer les fichiers d'image en fonction de leur extension
image_rdd = file_rdd.filter(lambda x: x[0].lower().endswith(('.jpg', '.jpeg', '.png', '.bmp')))

# Créer un RDD de lignes (Row) avec le contenu binaire du fichier et le chemin du fichier
image_rows = image_rdd.map(lambda x: Row(file_content=base64.b64encode(x[1]).decode('utf-8'), file_path=x[0]))

# Définir le schéma du DataFrame
schema = StructType([
    StructField("file_content", StringType(), nullable=True),
    StructField("file_path", StringType(), nullable=True)
])

# Créer le DataFrame en utilisant le RDD de lignes (Row) et le schéma
image_df = spark.createDataFrame(image_rows, schema)

# Enregistrer le DataFrame en tant que fichier CSV
# image_df.write.csv("/chemin/vers/le/fichier.csv", header=True)
# Enregistrer le DataFrame résultant dans un fichier CSV
# image_df.write.csv("C:\\Users\\Zbook\\OpenClassRoom\\ProjetsGitsOCR\\Projet_8\\data\\Result\\resultat.csv", header=True)


# print("Le DataFrame a été enregistré en tant que fichier CSV.")

# Affiche le nombre d'images chargées
print("Loaded {} images".format(image_df.count()))
```
